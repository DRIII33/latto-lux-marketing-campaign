```
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

# Authenticate with Google Cloud
auth.authenticate_user()

# Set up the BigQuery client
project_id = 'driiiportfolio'  # Your BigQuery Project ID
client = bigquery.Client(project=project_id)

# SQL query to pull the data from our combined view
sql_query = """
SELECT *
FROM `driiiportfolio.latto_lux_campaign.lttx_combined_view`
"""

# Run the query and store the results in a Pandas DataFrame
df = client.query(sql_query).to_dataframe()

# Display the first few rows of the DataFrame to verify
print("DataFrame Head:")
print(df.head())

# Basic EDA: Data Summary
print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics for numerical columns:")
print(df.describe())

# EDA Questions to Answer:
# 1. Which cities have the most ticket sales?
city_sales = df['retailer_location_city'].value_counts()
print("\nSales by City:")
print(city_sales)

# 2. What are the most popular app features?
feature_usage = df['feature_used'].value_counts()
print("\nApp Feature Usage:")
print(feature_usage)

# 3. How does app session length vary by city?
session_by_city = df.groupby('retailer_location_city')['app_session_length_minutes'].mean().sort_values(ascending=False)
print("\nAverage App Session Length by City:")
print(session_by_city)

# 4. What is the total number of social shares?
total_shares = df['social_share_count'].sum()
print(f"\nTotal Social Shares: {total_shares}")

# You can save this notebook and the output for the project's repository under the notebooks/ directory.
```

### EDA Summary

Based on the exploratory data analysis, we found the following key insights:

*   **Ticket Sales by City**: Houston has the highest number of ticket sales, followed by Atlanta and Los Angeles.
*   **App Feature Usage**: The most popular app features are "stream_music", "play_ar_game", and "scan_ticket". All features have relatively similar usage counts.
*   **App Session Length by City**: The average app session length is fairly consistent across all cities, with slight variations. Houston shows the highest average session length.
*   **Total Social Shares**: The total number of social shares generated through the app is 30011.

These findings provide a preliminary understanding of the campaign's performance and user engagement. Further analysis and visualization will be conducted to gain deeper insights and support decision-making.
